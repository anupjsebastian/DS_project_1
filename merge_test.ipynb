```python
def merge_movie_data(csv1, csv2):
    data1 = pd.read_csv(csv1)
    data2 = pd.read_csv(csv2)
    merged_df = pd.merge(data1, data2, on = 'Title', how = 'outer')
    return merged_df
```

```python
csv1 = 'movies.csv'
csv2 = 'allTime.csv'
```

```python
merged_A = merge_movie_data(csv1, csv2)
```
